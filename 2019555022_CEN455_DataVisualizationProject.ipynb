# Graph 1: Correlation Matrix

```python
import pandas as pd
import plotly.express as px

url = 'https://raw.githubusercontent.com/nerdropy/red-wine-visualization/master/winequality-red.csv'
df = pd.read_csv(url)

df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)

correlation_matrix = df.corr()
fig = px.imshow(correlation_matrix, labels=dict(x="Variables", y="Variables", color="Correlation"),
                x=correlation_matrix.index, y=correlation_matrix.columns, color_continuous_scale="redor")
fig.update_layout(title='Correlation Matrix')
fig.show()
```

# Graph 2: Strip Plot

```python
import pandas as pd
import plotly.express as px

url = 'https://raw.githubusercontent.com/nerdropy/red-wine-visualization/master/winequality-red.csv'
df = pd.read_csv(url)

df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)

fig = px.strip(df, x='quality', y='residual sugar',
               title='Residual Sugar Content Across Wine Qualities',
               labels={'residual sugar': 'Residual Sugar', 'quality': 'Quality'})
fig.show()
```

# Graph 3: Box Plot

```python
import pandas as pd
import plotly.express as px

url = 'https://raw.githubusercontent.com/nerdropy/red-wine-visualization/master/winequality-red.csv'
df = pd.read_csv(url)

df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)

fig = px.box(df, x='quality', y='total sulfur dioxide',
             title='Total Sulfur Dioxide Across Wine Qualities',
             labels={'total sulfur dioxide': 'Total Sulfur Dioxide', 'quality': 'Quality'})
fig.show()
```

#Graph 4: Scatter Plot

```python
import pandas as pd
import plotly.express as px

url = 'https://raw.githubusercontent.com/nerdropy/red-wine-visualization/master/winequality-red.csv'
df = pd.read_csv(url)

df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)

fig = px.scatter(df, x='density', y='quality',
                 title='Density vs. Wine Quality', labels={'density': 'Density', 'quality': 'Quality'})
fig.show()
```

#Graph 5: Violin Plot

```python
import pandas as pd
import plotly.express as px

url = 'https://raw.githubusercontent.com/nerdropy/red-wine-visualization/master/winequality-red.csv'
df = pd.read_csv(url)

df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)

fig = px.violin(df, x='quality', y='pH', box=True, points="all", hover_data=df.columns)
fig.update_layout(title='Variation of pH Across Wine Qualities',
                  xaxis_title='Quality',
                  yaxis_title='pH')
fig.show()
```
